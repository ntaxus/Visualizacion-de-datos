# *Wordcloud*

El siguiente es un código escrito en python para realizar un *wordcloud* a partir de los comentarios en un video del cantante Eminem publicado en la plataforma YouTube. El [set de datos](https://archive.ics.uci.edu/ml/machine-learning-databases/00380/) fue obtenido del repositorio de datos de la Universidad de California en Irvine.  


---

```python
# Python program to generate WordCloud
 
# importing all necessary modules
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
 
# leemos el file con los datos. En este caso usaremos los comentarios del file: 'Youtube04-Eminem.csv'
df = pd.read_csv(r"/media/nicolas/CONCOLOR/NICOLAS/BACKUP2022/MAESTRÍA CIENCIA DATOS/SEMESTRE 6/Visualizacion de datos/PEC2/YouTube-Spam-Collection-v1/Youtube04-Eminem.csv", encoding ="latin-1")
 
comment_words = ''
stopwords = set(STOPWORDS)
 
# iteramos a lo largo del archivo
for val in df.CONTENT:
     
    # casteamos a string
    val = str(val)
 
    # tokenizamos
    tokens = val.split()
     
    # pasamos cada token a lowcase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
     
    comment_words += " ".join(tokens)+" "
 
wordcloud = WordCloud(width = 1500, height = 1000,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
 
# plot the WordCloud image                      
plt.figure(figsize = (12, 12), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 10)
 
plt.show()
```
